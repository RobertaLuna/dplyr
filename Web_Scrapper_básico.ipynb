## Utilizando os verbos - GET

```python
# Escolha de site simples
site = "https://en.wikipedia.org/wiki/Web_scraping"

# Get
html = requests.get(site)
print(html.text)
```

## Utiliza√ß√£o de Formata√ß√£o com Beautiful Soup

```python
# Coleta os dados do site
html = requests.get(site).content

# Formatando os dados
dados = BeautifulSoup(html, 'html.parser')
print(dados.prettify())
```
