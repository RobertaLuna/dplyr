# Notebook para básico de webScrapping com Python

## Pacotes

```python
from bs4 import BeautifulSoup
import requests
```

## Utilizando os verbos - GET

```python
# Escolha de site simples
site = "https://en.wikipedia.org/wiki/Web_scraping"

# Get
html = requests.get(site)
print(html.text)
```

## Utilização de Formatação com Beautiful Soup

```python
# Coleta os dados do site
html = requests.get(site).content

# Formatando os dados
dados = BeautifulSoup(html, 'html.parser')
print(dados.prettify())
```

Com isso podemos utilizar a variável criada em com o beatififul soup para buscar informações.
